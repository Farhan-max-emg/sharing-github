**Handling Outliers in "Previous CTC" Column**

In the data preprocessing workflow, it was observed that the "Previous CTC" column contained outliers. Outliers are data points that significantly differ from the rest of the data and can impact the accuracy of our analysis and models. To ensure data quality and the reliability of our regression model, we applied an outlier handling technique.

**Outlier Handling Method: IQR (Interquartile Range)**

The Interquartile Range (IQR) method was used to identify and handle outliers in the "Previous CTC" column. The IQR is calculated as the difference between the 75th percentile (Q3) and the 25th percentile (Q1) of the data. Data points that fall below Q1 - 1.5 * IQR or above Q3 + 1.5 * IQR are considered outliers and were subsequently addressed.

**Impact Assessment**

After handling the outliers, it's essential to assess the impact of this step on the dataset. We should check if the removal of outliers affects the distribution of "Previous CTC" and whether it has any implications for our regression model, especially in predicting the dependent variable, "CTC."

```python
# Calculate the IQR for Previous CTC
Q1 = df["Previous CTC"].quantile(0.25)
Q3 = df["Previous CTC"].quantile(0.75)
IQR = Q3 - Q1

# Define the threshold for outliers
lower_threshold = Q1 - 1.5 * IQR
upper_threshold = Q3 + 1.5 * IQR

# Identify and cap outliers
df.loc[df["Previous CTC"] < lower_threshold, "Previous CTC"] = lower_threshold
df.loc[df["Previous CTC"] > upper_threshold, "Previous CTC"] = upper_threshold
```

Here Linear regression is well-suited for regression tasks, as it predicts continuous numerical values, like salary amounts. Logistic regression and classification are used for binary or multi-class categorization tasks, which are not aligned with our salary prediction objective.

###  Simple Linear Regression (statsmodels):

```python
import statsmodels.api as sm

# Example with one independent variable (e.g., "Previous CTC")
X_simple = df["Previous CTC"]
X_simple = sm.add_constant(X_simple)  # Add a constant term for the intercept
y = df["CTC"]

# Fit the simple linear regression model
model = sm.OLS(y, X_simple).fit()

# Print the model summary
model.summary()
```

### ----------------------------------------------------------------------------------------------------------------------------------------------------
**Summary:**
In the above code, we performed simple linear regression using the statsmodels library. Specifically, we analyzed the relationship between 'Previous CTC' (independent variable) and 'CTC' (salary - the target variable). The code fits a linear regression model, including an intercept, and prints a summary of the regression results.

**Explanation:**
Simple linear regression aims to understand the linear relationship between one independent variable ('Previous CTC') and the target variable ('CTC'). In the code, we added a constant term to the independent variable for the intercept. The model summary includes crucial information such as coefficients, R-squared value, p-values, and more, providing insights into the strength and significance of the relationship.

Certainly, here's a concise summary of the key points from the output of the regression model using statsmodels:

- **R-squared (R-squared):** It is a measure of how well the model explains the variance in the dependent variable. An R-squared of 0.067 indicates that the model explains approximately 6.7% of the variance in salary.

- **F-statistic (F-statistic):** It tests the overall significance of the model. A high F-statistic (113.2) and a low p-value (1.41e-25) suggest that the model is statistically significant.

- **Coefficients (coef):** These represent the relationships between variables. "Previous CTC" has a coefficient of 0.4897, indicating a positive relationship with salary.

- **P-values (P-value):** The p-value for "Previous CTC" is very low (0.000), indicating high statistical significance.

- **Omnibus, Jarque-Bera, Skew, and Kurtosis:** These values suggest that the residuals' distribution may not be normal, and the residuals may have skewness and heavy tails.

- **Condition Number (Cond. No.):** It checks for multicollinearity. A high value (4.71e+05) implies potential multicollinearity issues among independent variables.

In summary, the model is statistically significant, "Previous CTC" is a significant predictor, but it explains a limited portion of salary variance. Residuals may not be normally distributed, and there could be multicollinearity.
